```python
dls = CollabDataLoaders.from_df(ratings2, item_name='title', bs=64)
dls.show_batch()
```

```python
learn = collab_learner(dls, n_factors=50, y_range=(0, 5.5))
```

```python
learn.fit_one_cycle(7, 5e-3, wd=0.1)
```

```python
learn.model
```
